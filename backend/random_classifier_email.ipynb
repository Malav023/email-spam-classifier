```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.model_selection import RandomizedSearchCV

def make_pipeline(classifier):
    return Pipeline([
        ('cleaner', FunctionTransformer(clean_text)),
        ('tfidf', TfidfVectorizer(
            stop_words=None,
            ngram_range=(1, 1),
            max_features=2000
        )),
        ('clf', classifier)  # always 'clf' so param grids stay consistent
    ])

# Define models + their search spaces
search_spaces = {
    'RandomForest': (
        RandomForestClassifier(class_weight='balanced'),
        {
            'clf__n_estimators': [100, 200, 300],
            'clf__max_depth': [None, 10, 20],
            'clf__min_samples_leaf': [1, 2, 4],
            'tfidf__max_features': [1000, 2000, 5000]
        }
    ),
    'LogisticRegression': (
        LogisticRegression(class_weight='balanced', max_iter=1000),
        {
            'clf__C': [0.01, 0.1, 1, 10],
            'clf__solver': ['lbfgs', 'saga'],
            'tfidf__max_features': [1000, 2000, 5000]
        }
    ),
    'LinearSVC': (
        LinearSVC(class_weight='balanced'),
        {
            'clf__C': [0.01, 0.1, 1, 10],
            'tfidf__max_features': [1000, 2000, 5000]
        }
    ),
}

# Run RandomizedSearchCV for each model
results = {}
for name, (classifier, param_dist) in search_spaces.items():
    print(f"\nTuning {name}...")
    search = RandomizedSearchCV(
        make_pipeline(classifier),
        param_distributions=param_dist,
        n_iter=10,
        cv=5,
        scoring='f1',          # good for imbalanced spam data
        n_jobs=-1,
        random_state=42
    )
    search.fit(x_train, y_train)
    results[name] = {
        'best_score': search.best_score_,
        'best_params': search.best_params_,
        'best_estimator': search.best_estimator_
    }
    print(f"{name} best F1: {search.best_score_:.4f}")

# Pick the best overall model
best_name = max(results, key=lambda k: results[k]['best_score'])
best_model = results[best_name]['best_estimator']
print(f"\nBest model: {best_name} with F1={results[best_name]['best_score']:.4f}")

# Save it
joblib.dump(best_model, 'D:\Complete_proj\Email_spam\model\saved_models\spam_model_best.joblib')
```

```python
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import make_scorer, f1_score
import joblib

spam_f1 = make_scorer(f1_score, pos_label=1)

search_spaces = {
    'ComplementNB': (
        ComplementNB(),
        {
            'clf__alpha': [0.01, 0.05, 0.1, 0.5, 1.0, 2.0, 5.0],
            'clf__norm': [True, False],
            'tfidf__ngram_range': [(1,1), (1,2), (1,3)],
            'tfidf__max_features': [2000, 5000, 10000, None],
            'tfidf__sublinear_tf': [True, False],
            'tfidf__min_df': [1, 2, 3],
            'tfidf__max_df': [0.85, 0.90, 0.95, 1.0],
        }
    ),
    'MultinomialNB': (
        MultinomialNB(),
        {
            'clf__alpha': [0.01, 0.05, 0.1, 0.5, 1.0, 2.0, 5.0],
            'clf__fit_prior': [True, False],
            'tfidf__ngram_range': [(1,1), (1,2), (1,3)],
            'tfidf__max_features': [2000, 5000, 10000, None],
            'tfidf__sublinear_tf': [True, False],
            'tfidf__min_df': [1, 2, 3],
            'tfidf__max_df': [0.85, 0.90, 0.95, 1.0],
        }
    ),
}

results = {}

for name, (classifier, param_dist) in search_spaces.items():
    print(f"\nTuning {name}...")
    
    search = RandomizedSearchCV(
        make_pipeline(classifier),
        param_distributions=param_dist,
        n_iter=30,
        cv=5,
        scoring=spam_f1,
        n_jobs=-1,
        random_state=42,
        error_score='raise',
        verbose=1
    )
    
    search.fit(x_train, y_train)
    
    results[name] = {
        'best_score': search.best_score_,
        'best_params': search.best_params_,
        'best_estimator': search.best_estimator_
    }
    
    print(f"  Best F1 : {search.best_score_:.4f}")
    print(f"  Params  : {search.best_params_}")

# Pick best overall
best_name = max(results, key=lambda k: results[k]['best_score'])
best_model = results[best_name]['best_estimator']

print(f"\nBest model : {best_name}")
print(f"Best F1    : {results[best_name]['best_score']:.4f}")

# Evaluate on test set
from sklearn.metrics import classification_report
y_pred = best_model.predict(x_test)
print("\nTest Set Report:")
print(classification_report(y_test, y_pred))  

# Save
joblib.dump(best_model, 'D:\Complete_proj\Email_spam\model\saved_models\spam_model_nb_best.joblib')
print("Model saved.")
```
